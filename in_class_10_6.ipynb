```python
# set up pandas
import pandas as pd
df = pd.read_csv('https://ds1002-resources.s3.amazonaws.com/data/ride_sharing.csv')
```

```python
df.head(2)
```

```python
# sort the df
# df.sort_values(by="column_name", ascending=True), ascending gives in order starting at 1
# gives of head and tail at the same time

df.sort_values(by="station_A_id", ascending=True)
```

```python
# view only records based upon a given value of a given column
# look for all rows where station_A_id is equal to 81

# to do this, use the .query() method against your dataframe as a parameter
# set the "column_name == XX"

df.query('user_gender == "Female"')
```

```python
# we can also filter for string values with .query()

df.query("station_A_id == 21")
```
